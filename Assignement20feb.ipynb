### Solution (1)

#### GET method:
* Used to request data from a specified resource.
* It remain in browser history.
* GET request can be cached.
* It is used to append form data in url in name or value pair.
* The length of GET will limited.
* It is good for those data which does not require any security.

#### POST method:
* Used to request data from form.
* Provide security to data.
* POST request do not remain in the browser history.
* No restriction on data length.
* POST request are never cached.

### Solution (2)

#### Why is request used in Flask?
When flask application handle a request, it create a Request object based on the environment it received from wsgi server. because a worker (thread , process, or coroutine depending on the server) handle only one request at a time, the request data can be considered as global to that worker during that request.

### Solution (3)

#### Why is redirect() used in Flask?
When redirect() is called, it returns response object and redirect the user to another target location with specified status code, location parameter is the url where response should be redirected.

#### What are templates in Flask? Why is the render_template() function used?
Template file in a directory called templates inside your flask_app directory. Flask looks for templates in the templates directory, which is called templates

render_template() function which can be used to render the external HTML file to be returned as response from view function.
